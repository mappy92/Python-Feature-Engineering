Linear models assume that the dependent variables X take a linear relationship with the dependent variable Y. If the assumption is not met, the model may show poor performance. In this recipe, we will learn how to visualize the linear relationships between X and Y.

```python
import pandas as pd
import numpy as np

# for plotting
import matplotlib.pyplot as plt
import seaborn as sns

# the dataset for the demo
from sklearn.datasets import fetch_california_housing

# for linear regression
from sklearn.linear_model import LinearRegression
```

```python
# load the the california House price data from scikit-learn

# this is how we load the california dataset from sklearn
california_dataset = fetch_california_housing() ## Dataset is a json with all key values with data and metadata

# create a dataframe with the independent variables
california = pd.DataFrame(california_dataset.data,
                      columns=california_dataset.feature_names)

# add the target
california['MEDV'] = california_dataset.target

california.head()
```

```python
# this is the information about the california house prince dataset
# get familiar with the variables before continuing with 
# the notebook

# the aim is to predict the "Median value of the houses"
# MEDV column of this dataset

# and we have variables with characteristics about
# the homes and the neighborhoods

print(california_dataset.DESCR)
```

```python
# I will create a dataframe with the variable x that
# follows a normal distribution and shows a
# linear relationship with y

# this will provide the expected plots
# i.e., how the plots should look like if the
# linear assumption is met

np.random.seed(29) # for reproducibility

n = 200 # in the book we pass directly 200 within brackets, without defining n
x = np.random.randn(n) # randn for normal distribution, generates 200 points for a standard normal distriibution
y = x * 10 + np.random.randn(n) * 2 # 10x + 2(random no from standard normal distibution)

data = pd.DataFrame([x, y]).T ## Take x and y arrays as rows and then transposes it to make them columns
data.columns = ['x', 'y']
data.head()
```

Linear relationships can be assessed by scatter plots.

```python
# for the simulated data

# this is how the scatter-plot looks like when
# there is a linear relationship between X and Y

sns.lmplot(x="x", y="y", data=data, order=1)
# order 1 indicates that we want seaborn to
# estimate a linear model (the line in the plot below)
# between x and y

plt.ylabel('Target')
plt.xlabel('Independent variable')
```

```python
# now we make a scatter plot for the california
# house price dataset

# we plot the variable  AveOccup - average number of household members
# vs the target MEDV (median value of the house)

sns.lmplot(x="AveOccup", y="MEDV", data=california, order=1) # MedInc, HouseAge ,AveRooms	AveBedrms	Population	AveOccup	Latitude	Longitude	MEDV
```

Although not perfect, the relationship is fairly linear.

```python
# now we plot MedInc 
# vs the target MEDV (median value of the house)

sns.lmplot(x="MedInc", y="MEDV", data=california, order=1)
```

Linear relationships can also be assessed by evaluating the residuals. Residuals are the difference between the value estimated by the linear relationship and the real output. If the relationship is linear, the residuals should be normally distributed and centered around zero.

```python
# SIMULATED DATA

# step 1: build a linear model
# call the linear model from sklearn
linreg = LinearRegression()

# fit the model
linreg.fit(data['x'].to_frame(), data['y']) # <class 'pandas.core.series.Series'> to <class 'pandas.core.frame.DataFrame'>
# LinearRegression.fit() (and most scikit-learn models) require X (features) to be 2D  like a DataFrame or 2D NumPy array.
# A pandas Series is 1D, so passing it directly causes errors.
# .to_frame() wraps the Series into a DataFrame (2D).

# step 2: obtain the predictions
# make the predictions
pred = linreg.predict(data['x'].to_frame())

# step 3: calculate the residuals
error = data['y'] - pred

# plot predicted vs real
plt.scatter(x=pred, y=data['y'])
plt.xlabel('Predictions')
plt.ylabel('Real value')
```

```python
# step 4: observe the distribution of the residuals

# Residuals plot
# if the relationship is linear, the noise should be
# random, centered around zero, and follow a normal distribution

# we plot the error terms vs the independent variable x
# error values should be around 0 and homogeneously distributed

plt.scatter(y=error, x=data['x'])
plt.ylabel('Residuals')
plt.xlabel('Independent variable x')
```

```python
# step 4: observe the distribution of the errors

# plot a histogram of the residuals
# they should follow a gaussian distribution
# centered around 0

sns.distplot(error, bins=30)
plt.xlabel('Residuals')
```

```python
# now we do the same for the variable MedInc of the california
# house price dataset from sklearn

# call the linear model from sklearn
linreg = LinearRegression()

# fit the model
linreg.fit(california['MedInc'].to_frame(), california['MEDV'])

# make the predictions
pred = linreg.predict(california['MedInc'].to_frame())

# calculate the residuals
error = california['MEDV'] - pred

# plot predicted vs real
plt.scatter(x=pred, y=california['MEDV'])
plt.xlabel('Predictions')
plt.ylabel('MEDV')
```

```python
# Residuals plot

# if the relationship is linear, the noise should be
# random, centered around zero, and follow a normal distribution

plt.scatter(y=error, x=california['MedInc'])
plt.ylabel('Residuals')
plt.xlabel('MedInc')
```

```python
# plot a histogram of the residuals
# they should follow a gaussian distribution
sns.distplot(error, bins=30)
```

For this particular case, the residuals are not centered around zero, but they are not homogeneously distributed across the values of MedInc. Bigger and smaller values of MedInc show higher residual values. In addition, we see in the histogram that the residuals do not adopt a strictly Gaussian distribution.
